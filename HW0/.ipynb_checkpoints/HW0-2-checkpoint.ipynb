```python
#!/usr/bin/env python
# coding: utf-8

import csv
import numpy as np
import matplotlib.pyplot as plt


with open("data_1.csv") as fo:
    data1 = csv.reader(fo)
    X = [[]]
    y = []
    for row in data1:
        if(row[0] == "x"): continue
        X.append([pow(float(row[0]), i) for i in range(0,10)])
        y.append([float(row[1])])
    X = X[1:]
    X = np.array(X)
    y = np.array(y)


train_n = int(len(X)*0.7)
train_X = X[:train_n]
train_y = y[:train_n]
test_X = X[train_n:]
test_y = y[train_n:]


plt.figure()
plt.scatter([x[1] for x in X],y)
plt.xlabel("x")
plt.ylabel("t")
plt.title("x-t scatter graph")
plt.ylim(-1.2,1.2)
#plt.legend()
plt.show()


learning_rate = 0.1
epoch = 1000
C = 0.000


def calc_pred(w, x):
    return np.matmul(x, w)

def calc_error(p, y):
    return np.add.reduce((p-y)*(p-y))/(2*len(p))


def calc_grad(w, x, t, C):
    #print(np.matmul(x,w).shape)
    #print(x.shape)
    #w_norm2 = np.add.reduce(w*w)
    dedw = np.asarray([(-(t-np.matmul(x, w))*x[wi]+C*w[wi]) for wi in range(len(w))])
    return dedw


#do gradient descent:
#print(np.add.reduce((np.asarray([0, 0, 0])-np.asarray([2, 2, 2]))*(np.asarray([0, 0, 0])-np.asarray([2, 2, 2])))/(2*3))
#model = x(100, 10)*w(10, 1) = pred(100,1)
node_in = 10
node_out = 1

stat = dict()

for lnC in range():
    W = np.random.randn(node_in, node_out) / np.sqrt(node_in / 2)
    for ep_i in range(1, epoch+1):
        p = calc_pred(W, train_X)
        for g_i in range(len(train_X)):
            W -= learning_rate*calc_grad(W, train_X[g_i], train_y[g_i][0], C)
        if(ep_i%50 == 0):
            print("training error now = {}".format(calc_error(p, train_y)))
    test_error = calc_error(calc_pred(W, test_X), test_y)
    print("final error on test data = {}".format(test_error))
    
    

# In[185]:


plt.figure()
xs = np.linspace(0, 1, 1000)
ys = np.asarray([ np.matmul(np.asarray([[pow(x, i) for i in range(len(W))]]), W)[0][0] for x in xs])
print(np.asarray([x[1] for x in X]).shape)
print(np.asarray(y).shape)
plt.scatter([x[1] for x in X],y)
plt.plot(xs,ys)
plt.xlabel("x")
plt.ylabel("t")
plt.title("poly regression with M = {}, lr = {}, C = {}".format(9, learning_rate, C))
#plt.legend()
plt.show()


```
